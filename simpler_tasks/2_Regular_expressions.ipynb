```python
news_text = brown.words(categories='news')
fdist = nltk.FreqDist(w.lower() for w in news_text)
modals = ['can', 'could', 'may', 'might', 'must', 'will','what', 'who', 'when', 'why']
for m in modals:
    print (m + ':', fdist[m], end=' ')
```

```python
cfd = nltk.ConditionalFreqDist((genre, word) for genre in brown.categories() for word in brown.words(categories=genre))
genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
cfd.tabulate(conditions=genres, samples=modals)
```

```python
modals = [w for w in modals if w.startswith('wh')]
cfd.tabulate(conditions=genres, samples=modals)
```

```python
# This code creates a ConditionalFreqDist object from the inaugural corpus using a generator expression. 
# It then calls the plot() method on the resulting object to generate a plot of the frequencies of the target words 'america' and 'citizen' across the different inaugural addresses.
cfd = nltk.ConditionalFreqDist(
    (target, fileid[:4])
    for fileid in inaugural.fileids()
    for w in inaugural.words(fileid)
    for target in ['america', 'citizen']
    if w.lower().startswith(target))
cfd.plot()
```
